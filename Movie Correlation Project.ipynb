```python
import pandas as pd
import numpy as np
import seaborn as sns

import matplotlib
import matplotlib.pyplot as plt
import matplotlib.mlab as mlab

plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8) #This adjusts the configuration of the plots that are created


# reading in the data

df = pd.read_csv(r'C:\Users\User\Documents\Home\Projects\movies.csv')
```

```python
#sample top data
df.head()
```

```python
#check for missing values

for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{}-{}%'.format(col, round(pct_missing*100)))
```

```python
#check for data types of the columns
print(df.dtypes)
```

```python
#update column datatype
df['budget'] = df['budget'].astype('int64')
df['gross'] = df['gross'].astype('int64')
```

```python
df.head()
```

```python
#set object to string and take first 4 letters to make new column
df['year_correct'] = df['released'].astype(str).str[:4]
df.head()
```

```python
#order data
df = df.sort_values(by=['gross'], inplace=False, ascending=False)

```

```python
#set new number of rows visible
pd.set_option('display.max_rows', None)
```

```python
#show any duplicates

df['company'].drop_duplicates().sort_values(ascending=False)

#delete duplicates in company

#df['company'] = df['company'].drop_duplicates()
```

```python
#scatter plot with budget vs gross

plt.scatter(x=df['budget'], y=df['gross'])

plt.title('budget vs gross')

plt.xlabel('Budget')

plt.ylabel('gross Earnings')



plt.show
```

```python
#plot budget vs gross using seaborn

sns.regplot(x='budget', y='gross', data=df, scatter_kws={"color": "red"}, line_kws={"color": "blue"})
```

```python
#check for correlation of numeric fields
#pearson correlation
df.corr(method='pearson')
#kendall
#df.corr(method='kendall')

#Spearman
#df.corr(method='spearman')
```

```python
correlation_matrix = df.corr(method='pearson')

sns.heatmap(correlation_matrix, annot=True)

plt.title('Correlation Matrix for Numeric features')

plt.xlabel('Movie Features')

plt.ylabel('Movie Features')

plt.show()
```

```python
#check correlation of none numeric
#df.head()
df_made_numeric = df.copy() 

for col_name in df_made_numeric.columns:
    if(df_made_numeric[col_name].dtype == 'object'):
        df_made_numeric[col_name] = df_made_numeric[col_name].astype('category')
        df_made_numeric[col_name] = df_made_numeric[col_name].cat.codes
        
df_made_numeric.head()    
```

```python
df.head()
```

```python
correlation_matrix = df_made_numeric.corr(method='pearson')

sns.heatmap(correlation_matrix, annot=True)

plt.title('Correlation Matrix for Numeric features')

plt.xlabel('Movie Features')

plt.ylabel('Movie Features')

plt.show()
```

```python
correlation_mat2 = df_made_numeric.corr(method='pearson')
correlation_pairs = correlation_mat2.unstack()
sorted_pairs = correlation_pairs.sort_values()
sorted_pairs

high_corr = sorted_pairs[(sorted_pairs) > 0.5]
high_corr
```
